# 3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?
# Dispersion indicates how spread out the data is. Variance measures the average squared deviations from the mean, while standard deviation is its square root, providing spread in the same units as the data.

# Example Dataset: [10, 20, 30, 40, 50]
data = [10, 20, 30, 40, 50]
mean_data = sum(data) / len(data)
variance = sum((x - mean_data) ** 2 for x in data) / len(data)
std_dev = variance ** 0.5
print("Variance:", variance)
print("Standard Deviation:", std_dev)

# 7. What is the interquartile range (IQR), and how is it used to detect outliers?
# IQR = Q3 - Q1, measures the middle 50% spread of data.
# Outliers: Below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR.
import numpy as np
data = np.array(data)
q1, q3 = np.percentile(data, [25, 75])
iqr = q3 - q1
outlier_low = q1 - 1.5 * iqr
outlier_high = q3 + 1.5 * iqr
print("IQR:", iqr)
print("Outlier Range:", outlier_low, "to", outlier_high)

# 9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
# Properties: Symmetrical, bell-shaped, mean = median = mode.
# Empirical Rule: 68% within 1 SD, 95% within 2 SD, 99.7% within 3 SD.
import scipy.stats as stats

y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.title("Normal Distribution")
plt.show()

# 10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.
# Example: Calls received at a call center.
# Formula: P(X=k) = (e^(-lambda) * lambda^k) / k!
from math import exp, factorial
lambda_ = 3  # average calls per hour
k = 5         # specific number of calls
poisson_prob = (exp(-lambda_) * lambda_**k) / factorial(k)
print("Probability of", k, "calls:", poisson_prob)
