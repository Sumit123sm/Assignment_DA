# Theoretical questions and answers

## Theoretical Questions and Answers

### 1. What is a random variable in probability theory?
A random variable is a variable that can take on different values based on the outcomes of a random phenomenon. It maps outcomes of a random process to numerical values.

### 2. What are the types of random variables?
Random variables are classified into two types:
- **Discrete Random Variables**: Take on a countable number of distinct values.
- **Continuous Random Variables**: Can take on an infinite number of possible values within a given range.

### 3. What is the difference between discrete and continuous distributions?
- **Discrete Distribution**: Deals with discrete random variables that take on specific values (e.g., rolling a die).
- **Continuous Distribution**: Deals with continuous random variables that take on any value within a range (e.g., heights of people).

### 4. What are probability distribution functions (PDF)?
A Probability Distribution Function (PDF) specifies the probability of a random variable taking on a specific value for discrete variables or the probability density for continuous variables.

### 5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- **PDF**: Describes the likelihood of a random variable taking a specific value (or range of values for continuous variables).
- **CDF**: Represents the cumulative probability that a random variable will take a value less than or equal to a specific point.

### 6. What is a discrete uniform distribution?
A discrete uniform distribution is a distribution in which all possible outcomes of a discrete random variable are equally likely. For example, rolling a fair die.

### 7. What are the key properties of a Bernoulli distribution?
- A Bernoulli distribution has two possible outcomes: success (1) and failure (0).
- It is characterized by a single parameter, \( p \), which represents the probability of success.

### 8. What is the binomial distribution, and how is it used in probability?
The binomial distribution describes the number of successes in \( n \) independent Bernoulli trials, each with the same probability \( p \) of success.

### 9. What is the Poisson distribution and where is it applied?
The Poisson distribution models the number of times an event occurs in a fixed interval of time or space, given a constant average rate and independence of occurrences.

### 10. What is a continuous uniform distribution?
A continuous uniform distribution has all outcomes equally likely within a given range, defined by the interval [a, b].

### 11. What are the characteristics of a normal distribution?
- Symmetrical and bell-shaped.
- Mean, median, and mode are equal.
- Determined by two parameters: mean (\( \mu \)) and standard deviation (\( \sigma \)).

### 12. What is the standard normal distribution, and why is it important?
The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. It is essential for standardizing data and calculating probabilities.

### 13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The CLT states that the sampling distribution of the sample mean will approach a normal distribution as the sample size increases, regardless of the population's distribution. It is critical for making inferences about population parameters.

### 14. How does the Central Limit Theorem relate to the normal distribution?
The CLT explains why many real-world phenomena tend to follow a normal distribution, even if the underlying population distribution is not normal.

### 15. What is the application of Z statistics in hypothesis testing?
Z statistics are used in hypothesis testing to determine if a sample mean significantly differs from a population mean when the population variance is known.

### 16. How do you calculate a Z-score, and what does it represent?
A Z-score is calculated as:
\[ Z = \\frac{x - \\mu}{\\sigma} \]
It represents how many standard deviations a data point is from the mean.

### 17. What are point estimates and interval estimates in statistics?
- **Point Estimate**: A single value estimate of a population parameter (e.g., sample mean).
- **Interval Estimate**: A range of values (e.g., confidence interval) within which the population parameter is likely to fall.

### 18. What is the significance of confidence intervals in statistical analysis?
Confidence intervals provide a range of values to estimate a population parameter, offering a level of confidence (e.g., 95%) that the parameter lies within the range.

### 19. What is the relationship between a Z-score and a confidence interval?
Z-scores correspond to confidence levels. For example, a Z-score of 1.96 corresponds to a 95% confidence interval.

### 20. How are Z-scores used to compare different distributions?
Z-scores standardize data, enabling comparisons between datasets with different means and standard deviations.

### 21. What are the assumptions for applying the Central Limit Theorem?
- Random sampling.
- Independent observations.
- Finite variance.

### 22. What is the concept of expected value in a probability distribution?
The expected value is the weighted average of all possible values a random variable can take, weighted by their probabilities.

### 23. How does a probability distribution relate to the expected outcome of a random variable?
The expected outcome is derived from the probability distribution by calculating the weighted average of all possible values.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, uniform, norm, zscore, kurtosis, skew
from scipy.stats import t, sem
```

```python


# 1. Generate a random variable and display its value
random_value = np.random.rand()
print("Random Value:", random_value)
```

```python

# 2. Generate a discrete uniform distribution and plot PMF
discrete_values = np.arange(1, 7)  # Example: rolling a 6-sided die
pmf_values = [1/len(discrete_values)] * len(discrete_values)
plt.bar(discrete_values, pmf_values)
plt.title("PMF of Discrete Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.show()
```

```python
# 3. Calculate the PDF of a Bernoulli distribution
p = 0.3  # Probability of success
bernoulli_pdf = bernoulli(p)
x = [0, 1]
y = bernoulli_pdf.pmf(x)
plt.bar(x, y)
plt.title("PDF of Bernoulli Distribution")
plt.xlabel("Outcomes")
plt.ylabel("Probability")
plt.show()
```

```python
# 4. Simulate a binomial distribution and plot its histogram
n, p = 10, 0.5
binomial_data = binom.rvs(n, p, size=1000)
plt.hist(binomial_data, bins=range(n + 2), align='left', rwidth=0.8)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5. Create and visualize a Poisson distribution
lambda_poisson = 4  # Average rate
poisson_data = poisson.rvs(lambda_poisson, size=1000)
plt.hist(poisson_data, bins=range(max(poisson_data)+2), align='left', rwidth=0.8)
plt.title("Poisson Distribution (lambda=4)")
plt.xlabel("Occurrences")
plt.ylabel("Frequency")
plt.show()
```

```python

# 6. Calculate and plot the CDF of a discrete uniform distribution
cdf_values = np.cumsum(pmf_values)
plt.step(discrete_values, cdf_values, where='post')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Cumulative Probability")
plt.show()
```

```python

# 7. Generate a continuous uniform distribution and visualize it
uniform_data = uniform.rvs(size=1000)
plt.hist(uniform_data, bins=20, density=True)
plt.title("Continuous Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

```python
# 8. Simulate data from a normal distribution and plot its histogram
mean, std_dev = 0, 1
normal_data = norm.rvs(mean, std_dev, size=1000)
plt.hist(normal_data, bins=20, density=True)
plt.title("Normal Distribution (mean=0, std_dev=1)")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

```python
# 9. Calculate Z-scores from a dataset and plot them
data = np.random.randint(50, 100, size=10)
z_scores = (data - np.mean(data)) / np.std(data)
plt.bar(range(len(data)), z_scores)
plt.title("Z-Scores of a Dataset")
plt.xlabel("Index")
plt.ylabel("Z-Score")
plt.show()
```

```python
# 10. Implement the Central Limit Theorem (CLT) for a non-normal distribution
def central_limit_theorem_demo(sample_size, num_samples):
    means = []
    for _ in range(num_samples):
        sample = np.random.exponential(scale=2, size=sample_size)
        means.append(np.mean(sample))
    return means

clt_data = central_limit_theorem_demo(30, 1000)
plt.hist(clt_data, bins=20, density=True)
plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Means")
plt.ylabel("Density")
plt.show()
```

```python
# 15. Simulate multiple samples from a normal distribution and verify CLT
samples = [np.mean(norm.rvs(size=30)) for _ in range(1000)]
plt.hist(samples, bins=20, density=True)
plt.title("CLT Verification with Normal Distribution")
plt.xlabel("Sample Means")
plt.ylabel("Density")
plt.show()
```

```python
# 16. Plot the standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=0, scale=1)
plt.plot(x, y)
plt.title("Standard Normal Distribution")
plt.xlabel("Z")
plt.ylabel("Density")
plt.show()
```

```python
# 17. Generate random variables and calculate probabilities using binomial distribution
n, p = 10, 0.5
rv = binom(n, p)
probabilities = rv.pmf(range(n + 1))
plt.bar(range(n + 1), probabilities)
plt.title("Binomial Distribution Probabilities")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

```python
# 18. Calculate the Z-score for a given data point
point = 75
data = np.random.randint(50, 100, size=100)
z_score = (point - np.mean(data)) / np.std(data)
print("Z-score of the data point:", z_score)
```

```python
# 19. Hypothesis testing using Z-statistics
sample_mean = np.mean(data)
pop_mean = 70
pop_std = 10
n = len(data)
z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
print("Z-Statistic:", z_stat)
```

```python
# 20. Create a confidence interval
data = np.random.normal(100, 15, 50)
confidence = 0.95
ci = t.interval(confidence, len(data) - 1, loc=np.mean(data), scale=sem(data))
print("Confidence Interval:", ci)
```

```python
# 21. Generate normal data and interpret confidence interval
normal_data = np.random.normal(50, 10, 100)
mean, sigma = np.mean(normal_data), np.std(normal_data)
ci = t.interval(confidence, len(normal_data) - 1, loc=mean, scale=sem(normal_data))
print("Mean:", mean, "Confidence Interval:", ci)
```

```python
# 22. Calculate and visualize PDF of a normal distribution
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x)
plt.plot(x, pdf)
plt.title("PDF of Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Density")
plt.show()
```

```python
# 23. Calculate and interpret the CDF of Poisson distribution
lambda_poisson = 3
poisson_data = poisson.cdf(range(10), lambda_poisson)
plt.step(range(10), poisson_data, where='post')
plt.title("CDF of Poisson Distribution")
plt.xlabel("Values")
plt.ylabel("Cumulative Probability")
plt.show()
```

```python
# 24. Simulate a random variable using continuous uniform distribution
uniform_data = uniform.rvs(size=1000)
expected_value = np.mean(uniform_data)
print("Expected Value of Uniform Distribution:", expected_value)
```

```python
# 25. Compare standard deviations of two datasets
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 15, 100)
sd1, sd2 = np.std(data1), np.std(data2)
print("Standard Deviation of Dataset 1:", sd1)
print("Standard Deviation of Dataset 2:", sd2)
plt.bar(["Dataset 1", "Dataset 2"], [sd1, sd2])
plt.title("Comparison of Standard Deviations")
plt.show()
```

```python
# 26. Calculate range and IQR of a dataset
data = np.random.normal(50, 10, 100)
data_range = np.max(data) - np.min(data)
iqr = np.percentile(data, 75) - np.percentile(data, 25)
print("Range:", data_range)
print("IQR:", iqr)
```

```python

# 27. Implement Z-score normalization
data = np.random.randint(50, 100, size=20)
z_normalized = (data - np.mean(data)) / np.std(data)
print("Z-Score Normalized Data:", z_normalized)
```

```python
# 28. Calculate skewness and kurtosis of a dataset
data = np.random.normal(0, 1, 1000)
data_skewness = skew(data)
data_kurtosis = kurtosis(data)
print("Skewness:", data_skewness)
print("Kurtosis:", data_kurtosis)
```
