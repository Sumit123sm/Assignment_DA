"""
Statistics Part 2 - Theory

1. What is hypothesis testing in statistics?

   Hypothesis testing is a statistical method used to make inferences about a population based on sample data. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), then using statistical tests to determine which hypothesis is more likely.


2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

   The null hypothesis (H0) is a statement that assumes no effect or no difference, while the alternative hypothesis (H1) suggests the presence of an effect or difference. Hypothesis testing aims to determine whether there is enough evidence to reject H0 in favor of H1.


3. What is the significance level in hypothesis testing, and why is it important?

   The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01. It helps control the likelihood of making Type I errors.

4. What does a P-value represent in hypothesis testing?

   The P-value represents the probability of obtaining results as extreme as the observed data, assuming the null hypothesis is true. A lower P-value indicates stronger evidence against H0.

5. How do you interpret the P-value in hypothesis testing?

   - If P-value < α: Reject the null hypothesis (H0), supporting H1.
   - If P-value ≥ α: Fail to reject the null hypothesis (H0), indicating insufficient evidence.

6. What are Type 1 and Type 2 errors in hypothesis testing?

   - Type I Error: Rejecting a true null hypothesis (false positive).
   - Type II Error: Failing to reject a false null hypothesis (false negative).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

   - One-tailed test: Tests for an effect in only one direction (e.g., greater or smaller).
   - Two-tailed test: Tests for an effect in both directions.

8. What is the Z-test, and when is it used in hypothesis testing?

   The Z-test is used to compare a sample mean to a population mean when the population variance is known and the sample size is large (n > 30).

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

   The Z-score is calculated as:
   
   Z = (sample_mean - population_mean) / (population_std / sqrt(n))
   
   It represents how many standard deviations the sample mean is away from the population mean.

10. What is the T-distribution, and when should it be used instead of the normal distribution?

    The T-distribution is used when the population variance is unknown and the sample size is small (n < 30). It has heavier tails than the normal distribution.

11. What is the difference between a Z-test and a T-test?

    A Z-test is used when the population variance is known and the sample size is large, while a T-test is used when the population variance is unknown and the sample size is small.


12. What is the T-test, and how is it used in hypothesis testing?

    A T-test compares sample means to determine if they are significantly different. Types include one-sample, independent, and paired T-tests.

13. What is a confidence interval, and how is it used to interpret statistical results?

    A confidence interval is a range of values around a sample statistic that likely contains the population parameter. It provides an estimate with a given confidence level (e.g., 95%).

14. What is the margin of error, and how does it affect the confidence interval?

    The margin of error quantifies the uncertainty in an estimate. A larger margin of error results in a wider confidence interval.

15. How is Bayes' Theorem used in statistics, and what is its significance?

    Bayes' Theorem updates the probability of an event based on prior knowledge. It is widely used in Bayesian inference and decision-making.

16. What is the Chi-square distribution, and when is it used?

    The Chi-square distribution is used in hypothesis tests for categorical data, such as the Chi-square goodness-of-fit and independence tests.

17. What is the Chi-square goodness-of-fit test, and how is it applied?

    This test determines whether observed categorical data follows a specific expected distribution.



18. What is the F-distribution, and when is it used in hypothesis testing?

    The F-distribution is used in variance comparison tests such as the F-test and ANOVA.

19. What is an ANOVA test, and what are its assumptions?

    ANOVA (Analysis of Variance) compares means of multiple groups. Assumptions include normality, independence, and equal variance.

20. What are the different types of ANOVA tests?

    - One-way ANOVA: Compares means across one independent variable.
    - Two-way ANOVA: Examines interactions between two independent variables.

21. What is the F-test, and how does it relate to hypothesis testing?

    The F-test compares variances to determine statistical significance, commonly used in ANOVA and regression analysis.
"""


```python
# Practical Implementation

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
import scipy.stats as stats

# Sample data
sample = [52, 55, 53, 58, 54, 51, 56, 57, 55, 54]
population_mean = 50
population_std = 5

# Calculate Z-score
sample_mean = np.mean(sample)
sample_size = len(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value
p_value = 1 - stats.norm.cdf(z_score)

# Print results
print(f"Z-score: {z_score:.3f}, P-value: {p_value:.5f}")
if p_value < 0.05:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean?
import numpy as np
import scipy.stats as stats

# Sample data
sample = [60, 62, 65, 61, 66, 59, 63, 64, 60, 62]
population_mean = 63
population_std = 4

# Calculate Z-score
sample_mean = np.mean(sample)
sample_size = len(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate P-value
p_value = stats.norm.sf(abs(z_score)) * 2  # Two-tailed test

print(f"Z-score: {z_score:.3f}, P-value: {p_value:.5f}")
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    plt.figure(figsize=(8,5))
    plt.plot(x, y, label="Standard Normal Distribution")

    # Mark Type I error region (alpha)
    critical_value = stats.norm.ppf(1 - alpha)
    plt.fill_between(x, y, where=(x > critical_value), color='red', alpha=0.3, label='Type I Error (Alpha)')

    # Mark Type II error region (beta)
    shifted_x = x - 2  # Simulating alternate hypothesis distribution
    shifted_y = stats.norm.pdf(shifted_x, 0, 1)
    plt.plot(shifted_x, shifted_y, label="Alternate Hypothesis Distribution", linestyle="dashed")

    plt.fill_between(shifted_x, shifted_y, where=(shifted_x < critical_value), color='blue', alpha=0.3, label='Type II Error (Beta)')

    plt.legend()
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.show()

visualize_errors()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    sem = stats.sem(sample)  # Standard error of the mean
    margin = sem * stats.t.ppf((1 + confidence) / 2., len(sample) - 1)
    return mean - margin, mean + margin

sample = [54, 55, 56, 53, 58, 57, 54, 52, 56, 55]
ci = confidence_interval(sample)

print(f"95% Confidence Interval: {ci}")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_dev = np.std(sample, ddof=1)
    z_score = stats.norm.ppf((1 + confidence) / 2)
    return z_score * (std_dev / np.sqrt(n))

sample = [54, 55, 56, 53, 58, 57, 54, 52, 56, 55]
moe = margin_of_error(sample)
print(f"Margin of Error: {moe:.3f}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_A, likelihood_B_given_A, prior_B):
    posterior_A_given_B = (likelihood_B_given_A * prior_A) / prior_B
    return posterior_A_given_B

# Example: Disease testing
prior_disease = 0.01
likelihood_positive_given_disease = 0.95
prior_positive = 0.05

posterior = bayes_theorem(prior_disease, likelihood_positive_given_disease, prior_positive)
print(f"Posterior Probability (Having disease given positive test): {posterior:.3f}")

```

```python
# 16. Implement an F-test using Python to compare the variances of two random samples
import numpy as np
import scipy.stats as stats

sample1 = np.random.normal(50, 5, 30)
sample2 = np.random.normal(50, 8, 30)

f_stat = np.var(sample1, ddof=1) / np.var(sample2, ddof=1)

p_value = stats.f.cdf(f_stat, len(sample1) - 1, len(sample2) - 1)

print(f"F-Statistic: {f_stat:.3f}, P-value: {p_value:.5f}")

```

```python
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import scipy.stats as stats

def check_anova_assumptions(*groups):
    for i, group in enumerate(groups, start=1):
        stat, p = stats.shapiro(group)
        print(f"Group {i} - Normality Test: W-statistic={stat:.3f}, P-value={p:.5f}")

    stat, p = stats.levene(*groups)
    print(f"Levene’s Test for Equal Variance: Statistic={stat:.3f}, P-value={p:.5f}")

group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(55, 5, 30)
group3 = np.random.normal(60, 5, 30)

check_anova_assumptions(group1, group2, group3)
```

```python
# 20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Create the data
data = pd.DataFrame({
    "Factor_A": np.repeat(["A1", "A2"], 15),
    "Factor_B": np.tile(["B1", "B2", "B3"], 10),
    "Response": np.random.normal(50, 5, 30)
})

# Fit the two-way ANOVA model
model = ols('Response ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)

# Visualize the interaction between the factors
sns.pointplot(x="Factor_A", y="Response", hue="Factor_B", data=data, markers=["o", "s", "D"], linestyles=["-", "--", "-."])
plt.title("Interaction between Factor_A and Factor_B")
plt.show()
```

```python
# Task 22: Perform one-way ANOVA test and visualize results with boxplots.
def one_way_anova():
    np.random.seed(0)
    group1 = np.random.normal(10, 2, 30)
    group2 = np.random.normal(15, 2, 30)
    group3 = np.random.normal(20, 2, 30)

    data = [group1, group2, group3]

    # Boxplot
    plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])
    plt.title('Boxplot of Groups')
    plt.ylabel('Values')
    plt.show()

    # ANOVA test
    f_stat, p_value = stats.f_oneway(group1, group2, group3)
    print(f'F-statistic: {f_stat}, P-value: {p_value}')
```

```python
# Task 24: Hypothesis test for population variance using Chi-square distribution.
def chi_square_variance_test():
    np.random.seed(2)
    data = np.random.normal(0, 1, 50)
    sample_variance = np.var(data, ddof=1)
    population_variance = 1
    chi_square_stat = (len(data) - 1) * sample_variance / population_variance

    p_value = 1 - stats.chi2.cdf(chi_square_stat, df=len(data) - 1)
    print(f'Chi-square Statistic: {chi_square_stat}, P-value: {p_value}')
```

```python

# Task 25: Perform a Z-test for comparing proportions.
def z_test_proportions():
    count = np.array([50, 30])
    nobs = np.array([200, 150])
    z_stat, p_value = ztest(count / nobs, value=0)
    print(f'Z-statistic: {z_stat}, P-value: {p_value}')
```

```python
# Task 26: Perform F-test for comparing variances.
def f_test_variances():
    np.random.seed(3)
    data1 = np.random.normal(0, 1, 50)
    data2 = np.random.normal(0, 2, 50)

    f_stat = np.var(data1, ddof=1) / np.var(data2, ddof=1)
    dfn = len(data1) - 1
    dfd = len(data2) - 1
    p_value = 1 - stats.f.cdf(f_stat, dfn, dfd)

    print(f'F-statistic: {f_stat}, P-value: {p_value}')
```
